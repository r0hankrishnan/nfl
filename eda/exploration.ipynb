If we standardize all columns and then compute the difference between teams that did vs did not make the playoffs, we can see which variables differ the most between classes.

First, let's create a function that will standardize a given column.

```python
def standardize(column):
    return (column - column.mean())/column.std()
```

We can create the standardized data set by copying our original data set and applying our function to the relevant columns.

```python
#Copy df
standardizedDf = df.copy()

#Drop unnecessary columns 
standardizedDf = standardizedDf.drop(["Playoff", "Team", "Year"], axis = 1)

#Apply standardization function
standardizedDf = standardizedDf.apply(standardize)
```

Let's check if everything worked properly. All columns in the data frame should have a mean of 0 (*won't be exact because of rounding error*).

```python
for col in standardizedDf.columns:
    if round(standardizedDf[col].mean(),5) > 0 or round(standardizedDf[col].mean(),5) < 0:
        print("fail")
    else:
        pass
```

Our test passed so it looks like it worked. Now we can add the "Playoff" column back and compare the group means. 

```python
standardizedDf["Playoff"] = df["Playoff"]
standardizedDf = standardizedDf.groupby("Playoff").mean()

standardizedDf.transpose()
```

Finally, we can calculate the difference between playoff and non-playoff teams for each variable. 

```python
#Create empty data frame
differences = pd.DataFrame({"stat":[], "difference":[]})

#Iteratively calculate differences and add to the data frame
ind = 0
for col in standardizedDf:
    diff = standardizedDf[col][1] - standardizedDf[col][0]
    newRow = pd.DataFrame({"stat":col, "difference":diff}, index = [ind])
    differences = pd.concat([differences, newRow])
    ind = ind + 1
    
differences.sort_values(by = "difference")
```

### Look at correlations

```python
dfNoTeam = df.copy().drop("Team", axis = 1)
correlations = dfNoTeam.corr()
correlations = correlations.loc[:, "Playoff"]
correlations = correlations.reset_index().rename(columns = {"index": "variable"})
correlations = correlations[(correlations["Playoff"] > 0.50) | (correlations["Playoff"] < -0.50)]
correlations.sort_values(by = "Playoff")
```
